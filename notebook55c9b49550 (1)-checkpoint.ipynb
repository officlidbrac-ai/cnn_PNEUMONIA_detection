```python
def sum_all(*args):
    s=0
    for i in args:
        s==i
        return s
        
```

```python
def sum_all(*dog):
    s=0
    for i in dog:
        s+=i
        return s 
        
```

```python
# create a function to get a product of a number 
def product_all(*args):
    p=1
    for i in args:
        p*=i
        return p
        
```

```python
names=["Harry", "john", "karina", "Saif"] # print those names which have r in it
for i in names:
    if "r" in i:
        print(i)
```

```python
list(filter(lambda i: "r" in i, names))
```

```python
# print those names which have len less than or equal to 4
list(filter(lambda i : len(i)<=4 , names))
```
