### CONDITIONAL PROB

```python
#a fair die is rolled, what is the probability of getting even number
#when we know that it was more than 3 
```

```python
#A: even number [2,4,6]
#B: number more than 3 [4,5,6]
#A int B : [4,6]
#P(B) = 3/6
#P(A int B) =2/6
#P(A/B)= P(A int B)/ p(B)
```

```python
import numpy as np 
```

```python
All_event=np.array([1,2,3,4,5,6])
A=All_event[All_event%2==0]
B=All_event[All_event>3]
A_int_B=np.intersect1d(A,B)
P_A_given_B=len(A_int_B)/len(B)
P_A_given_B
```
